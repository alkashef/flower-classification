## Helper functions

```python
def show_dataset(dataset, n=10):
    img = np.vstack((np.hstack((np.asarray(dataset[i][0]) for _ in range(n))) for i in range(len(dataset))))
    plt.imshow(img)
    plt.axis('off')
```

## Load the data

```python
means = [0.485, 0.456, 0.406]
std = [0.229, 0.224, 0.225]

augment = transforms.Compose([transforms.Resize((224,224)),
                              transforms.ColorJitter(hue=.02, saturation=.02),
                              transforms.RandomHorizontalFlip(),
                              transforms.RandomVerticalFlip(),
                              transforms.RandomRotation(30, resample=PIL.Image.BILINEAR)])

crop = transforms.Compose([transforms.Resize((224,224))])
```

```python
aug_data = datasets.ImageFolder(aug_dir, transform=augment)
show_dataset(aug_data, n=6)
```
